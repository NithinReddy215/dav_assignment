```python
# Ensure 'Date' and 'Time' columns exist before processing
if 'Date' in df.columns and 'Time' in df.columns:
    df['datetime'] = pd.to_datetime(df['Date'].astype(str) + ' ' + df['Time'].astype(str))
    df['hour'] = df['datetime'].dt.hour
    df['day_of_week'] = df['datetime'].dt.dayofweek  # Monday=0, Sunday=6
    df['month'] = df['datetime'].dt.month
else:
    print("Date or Time column missing!")

# Borough extraction function
def extract_borough(address):
    if pd.isna(address):  # Handle missing addresses
        return 'Unknown'
    address = str(address)
    if 'BK NY' in address:
        return 'Brooklyn'
    elif 'NY NY' in address:
        return 'Manhattan'
    elif 'BX NY' in address:
        return 'Bronx'
    elif 'NJ' in address:
        return 'New Jersey'
    else:
        return 'Other'

# Ensure 'PU_Address' and 'DO_Address' exist before applying function
if 'PU_Address' in df.columns and 'DO_Address' in df.columns:
    df['PU_borough'] = df['PU_Address'].apply(extract_borough)
    df['DO_borough'] = df['DO_Address'].apply(extract_borough)
else:
    print("PU_Address or DO_Address column missing!")

# Check for missing values
print("\nMissing values:\n", df.isnull().sum())

# Display first few rows
print("\nPreview of processed data:\n", df.head())
```

```python
# Convert datetime column (Ensure it exists)
if 'datetime' in df.columns:
    df['datetime'] = pd.to_datetime(df['datetime'])

    # Extract hour from datetime
    df['hour'] = df['datetime'].dt.hour

    # Count rentals per hour
    rentals_per_hour = df.groupby('hour').size().reset_index(name='count')

    # Plot rental demand trend per hour
    plt.figure(figsize=(12, 5))
    sns.lineplot(x='hour', y='count', data=rentals_per_hour, marker='o', color='b')

    # Improve readability
    plt.xlabel("Hour of Day")
    plt.ylabel("Total Bike Rentals")
    plt.title("Bike Rental Demand Trend Per Hour")
    plt.xticks(range(0, 24))  # Ensure all hours are shown
    plt.grid(True)

    plt.show()
else:
    print("Error: 'datetime' column is missing!")
```

```python
# Ensure 'datetime' exists
if 'datetime' in df.columns:
    df['datetime'] = pd.to_datetime(df['datetime'])

    # Count rentals per day
    rentals_over_time = df.groupby(df['datetime'].dt.date).size().reset_index(name='count')

    # Plot rentals over time
    plt.figure(figsize=(12, 5))
    plt.plot(rentals_over_time['datetime'], rentals_over_time['count'], marker='o', linestyle='-', color='b')

    plt.xlabel("Date")
    plt.ylabel("Total Rentals")
    plt.title("Daily Bike Rental Trend")
    plt.xticks(rotation=45)
    plt.grid(True)

    plt.show()

else:
    print("Error: 'datetime' column is missing!")
```

```python
# Check if 'holiday' column exists
if 'holiday' in df.columns:
    # Convert 'datetime' to datetime format if not already
    df['datetime'] = pd.to_datetime(df['datetime'])

    # Group by holiday status (0 = Non-Holiday, 1 = Holiday)
    holiday_rentals = df.groupby('holiday').size().reset_index(name='count')

    # Convert holiday values to labels
    holiday_rentals['holiday'] = holiday_rentals['holiday'].map({0: 'Non-Holiday', 1: 'Holiday'})

    #  Bar Plot: Rentals vs. Holiday Status
    plt.figure(figsize=(8, 5))
    sns.barplot(x='holiday', y='count', data=holiday_rentals, hue='holiday', dodge=False, legend=False, palette=['blue', 'red'])

    plt.xlabel("Day Type")
    plt.ylabel("Total Rentals")
    plt.title("Bike Rentals: Holidays vs. Non-Holidays")  # Removed emoji to avoid font issues
    plt.grid(axis='y', linestyle='--', alpha=0.5)

    plt.show()

else:
    print("Error: 'holiday' column is missing!")
```
